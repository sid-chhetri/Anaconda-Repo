```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

## Simple Graph

```python

x1 = np.array([2,4,6,8,10])
y1 = x1 * 2

x2 = x1
y2 = x2 ** 2

plt.plot(x1,x2,linestyle = 'solid', label = 'Not good', marker = 'o', color = 'g')
plt.plot(x2,y2,linestyle = 'dotted', label = 'Good', marker = '*', color = 'b')

plt.ylabel("Grades")
plt.xlabel("Study Time")

plt.title("First Plot")
plt.legend(fontsize =  18,loc = 1)
plt.savefig("myplot.png")
plt.grid()
plt.show()
print()
```

## Scatter Plot

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y,label = 'scatterplot1', color = 'b', marker = '*')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
df = pd.DataFrame({"c1": [11,12,13,14,15], "c2": [22,26,24,25,30]})
plt.scatter(df.c1,df.c2, label = 'pdscatter', color = 'r', marker = 'o')


plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```

## Bar Chart

```python
stock_names = ["GOOG","APPL","IBM","INTC","CSCO"]
stock_value = [1000,800,600,400,200]

x_pos = range(len(stock_names))
#x_pos = [1,6,0,2,3]
#x_pos_ticks = [4,2,0,3,1]

plt.bar(x_pos, stock_value, width = 0.35, label = "stock_plot", align = "center", color = 'b', alpha = 1.0)

plt.xticks(x_pos,stock_names)
plt.xlabel('Name')
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend()
plt.grid()
plt.show()
```

```python
stock_names = ["GOOG","APPL","IBM","INTC","CSCO"]
stock_value = [1000,800,600,400,200]

y_pos = range(len(stock_names))

plt.barh(y_pos, stock_value, height = 0.5, label = "stock_plot", align = "center", color = 'c1', alpha = 1.0)

plt.yticks(y_pos,stock_names)
plt.xlabel('Price')
plt.ylabel('Name')
plt.title('Stock Prices')
plt.legend()
plt.grid()
plt.show()
```

```python
stock_names = ["GOOG","APPL","IBM","INTC","CSCO"]
stock_value = [1000,800,600,400,200]
stock_revenue = [500,400,300,200,100]
bar_width = 0.35

x_pos = np.arange(len(stock_names))
#x_pos = [1,6,0,2,3]
#x_pos_ticks = [4,2,0,3,1]

plt.bar(x_pos, stock_value, width = bar_width , label = "Stock Value", align = "center", color = 'c1', alpha = 1.0)
plt.bar(x_pos+bar_width,stock_revenue,width = bar_width, label = "Stock Revenue", align = 'center',color = 'r', alpha = 1.0)

plt.xticks(x_pos,stock_names)
plt.xlabel('Name')
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend()
plt.grid()
plt.show()
```

## Pie Chart

```python
slices = [46,10,20,14]
activities = ['Agriculture', ' Transport', 'Banking', 'IT']
colors = ['c','y','r','b']

plt.pie(labels=slices, colors=colors, x=slices, explode=(0,0.3,0,0), autopct='%1.1f%%')

plt.legend(activities)
plt.title('Pie Chart')
plt.show()
```

## Histogram

```python
import random
population_ages = [random.randint(1,100) for i in range(25)]
bins = list(range(0,101,10))
print(population_ages)
print(bins)
plt.hist(population_ages,bins,histtype = 'bar', align = 'mid', rwidth = 0.7, color = 'r')

plt.xlabel("Age")
plt.ylabel("Bincount")
plt.title("Binning")
plt.grid()
plt.show()
```

## Box Plot

```python
value1 = [random.randint(1,100) for i in range(25)]
value2 = [random.randint(1,100) for i in range(25)]

print(pd.DataFrame(value1).describe())

n1 = np.array(value1)
p1 = np.percentile(n1,(25,50,75))
print(p1)

box_plot_data = [value1,value2]
plt.boxplot(box_plot_data,labels={'series','series2'})
plt.grid()
plt.show()
```
